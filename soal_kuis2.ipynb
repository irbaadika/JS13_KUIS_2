```python
import pandas as pd
import numpy as np

df = pd.read_excel('data/dataset_quiz.xlsx')

df.head(100)
```

```python
# CEK DATA YANG N/A ATAU KOSONG

df.isna().sum()
```

```python
# Menghilangkan data null
df.drop('id_anggota5', axis=1, inplace=True)

m_ketua = df['id_ketua'].mode()[0]
df['id_ketua'] = df['id_ketua'].fillna(m_ketua)

m_anggota1 = df['id_anggota1'].mode()[0]
df['id_anggota1'] = df['id_anggota1'].fillna(m_anggota1)

m_anggota2 = df['id_anggota2'].mode()[0]
df['id_anggota2'] = df['id_anggota2'].fillna(m_anggota2)

m_anggota3 = df['id_anggota3'].mode()[0]
df['id_anggota3'] = df['id_anggota3'].fillna(m_anggota3)

m_anggota4 = df['id_anggota4'].mode()[0]
df['id_anggota4'] = df['id_anggota4'].fillna(m_anggota4)
# Cek Data
df.head()
```

```python
# CEK DATA YANG N/A ATAU KOSONG

df.isna().sum()
```

```python
# MENENTUKAN KOLOM YANG MENJADI LABEL

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['keterlibatan_vltr'] = le.fit_transform(df['keterlibatan_vltr'])
df['skema'] = le.fit_transform(df['skema'])
df['luaran_hki'] = le.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki'] = le.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki'] = le.fit_transform(df['potensi_hki'])
df['tahun'] = le.fit_transform(df['tahun'])
df['luaran_lain'] = le.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain'] = le.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain'] = le.fit_transform(df['status_luaran_lain'])
df.head()
```

```python
# SELEKSI

x = df.iloc[:,6:]
y = df['keterlibatan_vltr']

x.shape
```

```python
# DATA TRAIN DAN DATA VALIDASI

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

x_val, x_unseen, y_val, y_unseen = train_test_split(x_test, y_test, test_size=0.6, random_state=1)
```

```python
# MENGGUNAKAN DECISION TREE

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# FIT dt
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)

# PREDICT dt
y_pred_dt = dt.predict(x_test)

# SET ACCURACY
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy: {:.2f}".format(accuracy_dt))
print(f"Test set accuracy: {accuracy_dt}")
```

```python
# MENGGUNAKAN RANDOM FOREST

from sklearn.ensemble import RandomForestClassifier

# FIT dt
rfc = RandomForestClassifier(n_estimators=20, random_state=1)
rfc.fit(x_train, y_train)

# Memprediksi label set test
y_pred_rfc = rfc.predict(x_test)

# Menghitung set accuracy
accuracy_rfc = accuracy_score(y_test, y_pred_rfc)
print("Test set accuracy: {:.2f}".format(accuracy_rfc))
print(f"Test set accuracy: {accuracy_rfc}")
```

```python
# GRIDSEARCH DENGAN DECISION TREE
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold

# INISIASI REPEATED K FOLD
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)

# HIPERPARAMETER
params = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'max_depth' : list(range(5,11))
}

# INISIASI GRID
grid = GridSearchCV(clf_tree, param_grid=params, cv=cv)

# FIT GRID
%timeit grid.fit(x,y)

# EVALUASI
score = grid.score(x,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
# RANDOM SEARCH DENGAN DECISION TREE
from sklearn.model_selection import RandomizedSearchCV

# HYPERPARAMMETER
params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

# INISIASI GRID
random = RandomizedSearchCV(clf_tree, params, cv=cv)

# FIT RANDOM
%timeit random.fit(x,y)

# EVALUASI
score = random.score(x,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {random.best_params_}')
```
